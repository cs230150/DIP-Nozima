```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/Users/230150/Downloads/nature.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

image_sub = cv2.subtract(img_rgb, 100)

plt.imshow(image_sub)
plt.title("Subtracted Image")
plt.axis('off')
plt.show()
```

When we subtract a bigger intensity value, pixel values decrease more, so the image becomes darker. Some pixels may become zero, which removes details.

```python
red_reduced = img_rgb.copy()
red_reduced[:,:,0] = cv2.subtract(red_reduced[:,:,0], 100)

plt.imshow(red_reduced)
plt.title("Red Channel Reduced")
plt.axis('off')
plt.show()
```

```python
add_30 = cv2.add(img_rgb, 30)
add_100 = cv2.add(img_rgb, 100)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1); plt.imshow(img_rgb); plt.title("Original"); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(add_30); plt.title("Added 30"); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(add_100); plt.title("Added 100"); plt.axis('off')
plt.show()
```

Adding intensity increases pixel values, so the image becomes brighter. Very high values may reach the maximum (255), causing over-brightness.

```python
red_added = img_rgb.copy()
red_added[:,:,0] = cv2.add(red_added[:,:,0], 1000)

plt.imshow(red_added)
plt.title("Red Channel Increased")
plt.axis('off')
plt.show()                    
```

```python
mul_half = cv2.multiply(img_rgb, 0.5)
mul_two = cv2.multiply(img_rgb, 2)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1); plt.imshow(img_rgb); plt.title("Original"); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(mul_half); plt.title(" 0.5"); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(mul_two); plt.title(" 2"); plt.axis('off')
plt.show()
```

Multiplying by a number greater than 1 increases brightness and contrast. Multiplying by a number less than 1 darkens the image and reduces contrast.

```python
div_2 = cv2.divide(img_rgb, 2)
div_05 = cv2.divide(img_rgb, 0.5)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1); plt.imshow(img_rgb); plt.title("Original"); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(div_2); plt.title(" 2"); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(div_05); plt.title(" 0.5"); plt.axis('off')
plt.show()
```

Dividing by a number greater than 1 reduces brightness. Dividing by a number less than 1 increases brightness, similar to multiplication.

In this lab, I learned how arithmetic operations affect image brightness and contrast. Subtraction and division darken images, while addition and multiplication brighten them. I also learned how to manipulate individual color channels, such as modifying only the red channel.
